```python
import pandas as pd
from sklearn.linear_model import LinearRegression
import matplotlib.pyplot as plt
import math
from word2number import w2n
```

```python
df = pd.read_csv('hiring.csv')
```

```python
df.info()
```

```python
df.head(10)
```

```python
df =df.rename(columns={'test_score(out of 10)':'test_score',
                   'interview_score(out of 10)':'interview_score',
                   'salary($)':'salary'})
```

```python
df.experience = df.experience.fillna('zero')
mean_score = math.floor(df['test_score'].mean())
df['test_score'] = df['test_score'].fillna(mean_score)
df
```

```python
df.experience = df.experience.apply(lambda x: w2n.word_to_num(x))
df
```

```python
lr = LinearRegression()
lr.fit(df[['experience','test_score','interview_score']],df.salary)
```

```python
M=lr.coef_
b=lr.intercept_
```

```python
lr.predict([[2,9,6],[12,10,10]])
```
